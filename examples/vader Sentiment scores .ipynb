```python
import numpy as np # linear algebra, post installation using pip install numpy
import pandas as pd # data processing, post installation using pip install pandas
import nltk # for sentiment analysis, VADER is the part of this package post installation using pip install nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
df = pd.read_csv("Twitter_Data.csv")
df.head()
```

```python

analyzer = SentimentIntensityAnalyzer()
df['scores'] = df['Tweet'].apply(lambda x: analyzer.polarity_scores(str(x)))
df.head()
```

```python
df['neg'] = df['Tweet'].apply(lambda x:analyzer.polarity_scores(str(x))['neg'])
df['neu'] = df['Tweet'].apply(lambda x:analyzer.polarity_scores(str(x))['neu'])
df['pos'] = df['Tweet'].apply(lambda x:analyzer.polarity_scores(str(x))['pos'])
df['compound'] = df['Tweet'].apply(lambda x:analyzer.polarity_scores(str(x))['compound'])
df.head()
```

```python
df['sentiment_label']=''
df.loc[df.compound>0,'sentiment_type']=1
df.loc[df.compound==0,'sentiment_type']=0
df.loc[df.compound<0,'sentiment_type']=-1
df.head()
```
